```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Example data generation (replace with your data)
x = np.random.normal(loc=0, scale=1, size=200)

# Plot histogram and KDE curve over it
sns.histplot(x, kde=True, stat="count", color="steelblue", edgecolor="black")

# Fit a normal distribution and overlay it
xmin, xmax = plt.xlim()
x_vals = np.linspace(xmin, xmax, 100)
mean, std = norm.fit(x)
plt.plot(x_vals, norm.pdf(x_vals, mean, std) * len(x) * (xmax - xmin)/len(x_vals), color='black', linewidth=2)

plt.show()
```
