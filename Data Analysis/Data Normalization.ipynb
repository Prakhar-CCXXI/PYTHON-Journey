```python
import pandas as pd 
path = "C:/Users/PRAKHAR/AppData/Local/Programs/Kaggle/Test.xlsx"
df = pd.read_excel(path)
print(df)
```

```python
df.head(5)
```

```python
# SIMPLE FEATURE SCALING METHOD 
df["B"]=df["B"]/df["B"].max()
```

```python
df.head(5)
```

```python
# MIN-MAX METHOD
df["B"]=(df["B"]-df["B"].min())/(df["B"].max()-df["B"].min())
```

```python
df.head(5)
```

```python
# Z-SCORE METHOD
df["B"]=(df["B"]-df["B"].mean())/df["B"].std()
```

```python
df.head(5)
```

```python
# NUMERIC TO CATEGORICAL 
import numpy as np
bins = np.linspace(min(df["B"]),max(df["B"]),4)
```

```python
group_names = ["Low","Medium","High"]
```

```python
df["B-binned"] = pd.cut(df["B"],bins,labels = group_names , include_lowest = True)
```

```python
df.head(5)
```

```python
# CATEGORICAL TO NUMERIC
import pandas as pd
pd.get_dummies(df["B"] , dtype=int)
```
